```python
# === Imports ===
from langchain_community.document_loaders import DirectoryLoader, PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
import os
import time
import traceback

print("ðŸš€ Starting FAISS embedding pipeline with detailed debugging...\n")

try:
    # === STEP 1: Load all PDFs from folder ===
    start_time = time.time()
    dir_path = "../data/pdf"
    print(f"ðŸ“‚ Looking for PDFs in: {os.path.abspath(dir_path)}")

    dir_loader = DirectoryLoader(
        dir_path,
        glob="**/*.pdf",
        loader_cls=PyMuPDFLoader,
        show_progress=True
    )
    pdf_documents = dir_loader.load()
    print(f"âœ… Loaded {len(pdf_documents)} total pages in {time.time() - start_time:.2f}s")

    if not pdf_documents:
        raise FileNotFoundError("No PDF documents found! Check your path or file names.")

    # === STEP 2: Show metadata for each file ===
    print("\nðŸ“‘ Extracting unique PDF metadata...")
    unique_files = {}
    for doc in pdf_documents:
        src = doc.metadata.get("source", "unknown")
        if src not in unique_files:
            unique_files[src] = doc.metadata

    print(f"ðŸ“„ Found {len(unique_files)} unique PDF files:")
    for i, (src, meta) in enumerate(unique_files.items(), 1):
        print(f"   âž¤ {i}. {src}")
        for key, value in meta.items():
            print(f"       - {key}: {value}")
        print("-" * 80)

    # === STEP 3: Split documents into chunks ===
    print("\nðŸ”ª Splitting documents into text chunks...")
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=150
    )
    texts = text_splitter.split_documents(pdf_documents)
    print(f"âœ… Split into {len(texts)} text chunks")

    # Debug sample chunk
    print("\nðŸ§© Sample chunk preview:")
    for i, doc in enumerate(texts[:2], 1):
        print(f"\n--- Chunk {i} (len={len(doc.page_content)}) ---")
        print(doc.page_content[:300], "...")
        print(doc.metadata)

    # === STEP 4: Create Ollama embeddings (nomic-embed-text:latest) ===
    print("\nðŸ§  Initializing Ollama Embeddings (nomic-embed-text:latest)...")
    embeddings = OllamaEmbeddings(model="nomic-embed-text:latest")
    print("âœ… Embedding model loaded successfully")

    # === STEP 5: Build FAISS Vectorstore ===
    print("\nâš™ï¸ Creating FAISS vectorstore (this may take a while)...")
    start_time = time.time()
    vectorstore = FAISS.from_documents(texts, embeddings)
    print(f"âœ… FAISS vectorstore created in {time.time() - start_time:.2f}s")

    # Debug FAISS stats
    index_size = len(texts)
    print(f"ðŸ“Š Vectorstore contains {index_size} vectors (dimension: {len(embeddings.embed_query('test'))})")

    # === STEP 6: Save FAISS index ===
    faiss_dir = "faiss_db_nomic"
    os.makedirs(faiss_dir, exist_ok=True)
    vectorstore.save_local(faiss_dir)
    print(f"ðŸ’¾ FAISS vectorstore saved successfully at: {os.path.abspath(faiss_dir)}")

    # === STEP 7: Optional reload verification ===
    print("\nðŸ” Reloading saved FAISS store for validation...")
    reloaded = FAISS.load_local(faiss_dir, embeddings, allow_dangerous_deserialization=True)
    print("âœ… Reloaded FAISS vectorstore verified successfully.")

    # Simple test query
    print("\nðŸ’¬ Running a quick similarity test:")
    query = "Explain the key idea of the document"
    docs = reloaded.similarity_search(query, k=2)
    for i, d in enumerate(docs, 1):
        print(f"\nðŸ”¹ Result {i}:")
        print(d.page_content[:300], "...")
        print(f"Metadata: {d.metadata}")

except Exception as e:
    print("\nâŒ ERROR OCCURRED:")
    traceback.print_exc()
```

```python
# === Imports ===
import os
import time
from langchain_community.llms import Ollama
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.chains import RetrievalQA

# Correct import for PromptTemplate
from langchain.prompts import PromptTemplate

# Correct import for RetrievalQA


# --- Configuration ---
LLM_MODEL = "qwen2:0.5b"
EMBEDDING_MODEL = "nomic-embed-text:latest"
FAISS_DIR = "faiss_db_nomic"

# --- Step 1: Load embeddings ---
embeddings = OllamaEmbeddings(model=EMBEDDING_MODEL)

# --- Step 2: Load FAISS ---
vectorstore = FAISS.load_local(FAISS_DIR, embeddings, allow_dangerous_deserialization=True)

# --- Step 3: LLM & Prompt ---
llm = Ollama(model=LLM_MODEL)

template = """You are an accurate, helpful AI assistant. Answer the user's question based ONLY on the context provided below. 
If you cannot find the answer in the context, clearly state that the answer is not available in the documents.

Context:
{context}

Question: {question}

Answer:"""

prompt = PromptTemplate(template=template, input_variables=["context", "question"])

retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt}
)

# --- Step 4: Query ---
query = "What are the core principles of cryptography as discussed in the documents?"
result = qa_chain.invoke({"query": query})

print("\n--- ðŸ¤– Answer ---")
print(result['result'])

print("\n--- ðŸ“‘ Sources ---")
sources = set([doc.metadata.get('source', 'N/A') for doc in result['source_documents']])
for source in sources:
    print(f"ðŸ”¹ Source: {source}")
```
